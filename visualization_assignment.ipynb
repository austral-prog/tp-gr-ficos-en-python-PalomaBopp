```python
import matplotlib.pyplot as plt
```

```python
dias = [1, 2, 3, 4, 5, 6, 7]  # Eje X: días de la semana
temperaturas = [22, 24, 20, 23, 25, 21, 22]  # Eje Y: temperatura en grados

# Graficar
plt.plot(dias, temperaturas, marker='o', color='violet', linestyle='-')
# Etiquetas
plt.title('Temperatura durante la semana')
plt.xlabel('Día')
plt.ylabel('Temperatura (°C)')
```

### Representa la relacion entre la temperatura y los dias de la semana.La eleccion fue porque se ve claramente como decrece y aumenta la temperatura

```python
materias = ['Matemática', 'Física', 'Programación', 'Inglés', 'Historia']
notas = [8, 7, 9, 6, 7]

# Crear figura más grande
fig, ax = plt.subplots(figsize=(10, 6))  # Tamaño ampliado

# Graficar barras
ax.bar(materias, notas, color='pink')

# Etiquetas
ax.set_title('Notas finales por materia')
ax.set_xlabel('Materia')
ax.set_ylabel('Nota (0-10)')
```

### Mustra la relacion entre las notas finales y las materias rendidas.La eleccion fue que se ve la diferencia entre las materias de una manera clara.

```python
horas_estudio = [1, 2, 3, 4, 5, 6, 7]  # Eje X: horas de estudio por día
notas = [4, 5, 6, 6.5, 7, 8.5, 9]      # Eje Y: nota obtenida en un examen

# Graficar puntos
plt.scatter(horas_estudio, notas, color='green')

# Etiquetas
plt.title('Relación entre horas de estudio y nota')
plt.xlabel('Horas de estudio')
plt.ylabel('Nota en examen')
```

### Relacion entre horas de estudio y su dia, es decir cuanto se estudio por dia.La eleccion fue que por cada punto en particular al representar la relacion es mas claro para ver.
